# Coffee Production Web Scraping Project

Importing our needed library's

```python
#Make sure that the first letter
from bs4 import BeautifulSoup
import requests as re 
```

Declare a variable that contains the URL we want to extract the data from.

```python
url='https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production'
data= re.get(url)
```

```python
soup=BeautifulSoup(data.text)
```

```python
table_value=soup.find('table')
```

```python
table_value
```

```python
for i in table_value:
    print(i.text)
```

```python
# as data is not organised, we want to get table row one by one, tr means table rows
ls=[]
for i in table_value.find_all('tr'):
    raw_data=i.text
    data=raw_data.replace('\n\n','-').strip()
    ls.append(data.split('-'))
ls    
```

```python
import pandas as pd 
df= pd.DataFrame(ls[1:],columns=ls[0])
df
```

```python
df.info()
```

```python
df['60 kilogram bags'] = df['60 kilogram bags'].str.replace(',','') 
df['60 kilogram bags'] = df['60 kilogram bags'].astype(float)
```

```python
df['Metric tons']=df['Metric tons'].str.replace(',','')
df['Metric tons']=df['Metric tons'].astype(float)

df['Pounds']=df['Pounds'].str.replace(',','')
df['Pounds']=df['Pounds'].astype(float)
```
